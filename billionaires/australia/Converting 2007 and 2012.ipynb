```python
import pandas as pd
import numpy as np
```

```python
Part12007 = pd.read_csv("2007 Table 1.csv", header = None )
Part12007
```

```python
one = Part12007.copy().loc[:, 0:1]
two = Part12007.copy().loc[:, 2:4]
three = Part12007.copy().loc[:, 5:7]
```

```python
one["Ranking"] = 0
```

```python
for i in one.index:
    ranks = one.loc[i,0].split(" ")
    if i == 49:
        one.loc[i,0] = one.loc[i,0][4:]
    else:
        one.loc[i,0] = one.loc[i,0][3:]
    one.loc[i, "Ranking"] = ranks[0]
```

```python
one.columns = ["Name", "Money", "Rank"]
two.columns = ["Rank", "Name", "Money"]
three.columns = ["Rank", "Name", "Money"]
```

```python
pd.concat([one, two, three])
```

```python
Part22007 = pd.read_csv("2007 Table 2.csv", header = None )
Part22007
```

```python
new = Part22007.copy().loc[:, 0:2]
new.columns = ["Rank", "Name", "Money"]
final = pd.concat([new, one, two, three])
final.to_csv("Final 2007.csv", index = False)
```

```python
Part12012 = pd.read_csv("2012 Table 1.csv", header = None )
Part22012 = pd.read_csv("2012 Table 2.csv", header = None )
Part32012 = pd.read_csv("2012 Table 3.csv", header = None )
Part42012 = pd.read_csv("2012 part 2.csv" )
new_one = Part12012.copy().loc[:, 0:2]
new_two = Part22012.copy().loc[:, 0:2]
new_three = Part32012.copy().loc[:, 0:2]
new_four = Part42012.copy().iloc[:, 0:3]
```

```python
new_one.columns = ["Rank", "Name", "Money"]
new_two.columns = ["Rank", "Name", "Money"]
new_three.columns = ["Rank", "Name", "Money"]
new_four.columns = ["Rank", "Name", "Money"]
```

```python
for i in new_one.index:
    new_one.loc[i, "Money"] = new_one.loc[i, "Money"].replace("$", "").replace(",", "")
for i in new_two.index:
    new_two.loc[i, "Money"] = new_two.loc[i, "Money"].replace("$", "").replace(",", "")
for i in new_three.index:
    
    try:
        new_three.loc[i, "Money"] = new_three.loc[i, "Money"].replace("$", "").replace(",", "")
    except:
        print(i)
for i in new_four.index:
    new_four.loc[i, "Money"] = new_four.loc[i, "Money"].replace("$", "").replace(",", "")



```

```python
for i in new_three.index:
    if np.isnan(new_three.loc[i, "Rank"]):
        new_three.loc[i-1, "Name"] = new_three.loc[i-1, "Name"] + " " + new_three.loc[i, "Name"]
```

```python
new_three= new_three.dropna()
```

```python
Final_2012 = pd.concat([new_one, new_two, new_three, new_four])
```

```python
Final_2012["Rank"] = Final_2012["Rank"].astype(int)
Final_2012
```

```python
Final_2012.sort_values(by=['Rank'], inplace = True)
```

```python
Final_2012.to_csv("Final_2012.csv", index = False)
```
