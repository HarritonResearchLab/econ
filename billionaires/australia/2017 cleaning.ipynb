```python
bios = []
between.append(len(file))
for i in range(len(between)-1):
    bios.append(file[between[i]:between[i+1]])
```

```python
results = pd.DataFrame(columns = ["Rankings", "Name", "2017 Wealth", "2016 Wealth", "Sector", "Age", "Location", "Additional Text"], index = [i for i in range(len(bios))])

for a,i in enumerate(bios):
    comma = False
    if i[2] == "," or i[3] == ",":
        comma = True
    if i[0].isnumeric() and not i[1].isnumeric():
        results.loc[a, "Rankings"] = i[0]
        results.loc[a, "Name"] =''.join([i for i in  i[2:i.find("$")] if not i.isdigit()]).replace(',', '')
    elif i[:3].isnumeric():
        results.loc[a, "Rankings"] = i[:3]
        results.loc[a, "Name"] = ''.join([i for i in i[3:i.find("$")] if not i.isdigit()]).replace(',', '')
    elif i[:2].isnumeric():
        results.loc[a, "Rankings"] = i[:2]
        results.loc[a, "Name"] = ''.join([i for i in i[2:i.find("$")] if not i.isdigit()]).replace(',', '')
        
    new1 = i[i.find("$"):]
    x = new1[:new1.find("n")+1].replace(" ", "")
    if x[1:6].replace(".", "").isnumeric():
        results.loc[a, "2017 Wealth"] = int(float(x[1:6]) * 1000)
    elif x[1:5].replace(".", "").isnumeric():
        results.loc[a, "2017 Wealth"] = int(float(x[1:5]) * 1000)
    elif x[1:4].replace(".", "").isnumeric():
        results.loc[a, "2017 Wealth"] = int(float(x[1:4]))
        
    new1 = new1[1:]
    new2 = new1[new1.find("$"):]
    x = new2[:new2.find("N")+1].replace(" ", "")
    if x[1:6].replace(".", "").isnumeric():
        results.loc[a, "2016 Wealth"] = int(float(x[1:6]) * 1000)
    elif x[1:5].replace(".", "").isnumeric():
        results.loc[a, "2016 Wealth"] = int(float(x[1:5]) * 1000)
    elif x[1:4].replace(".", "").isnumeric():
        results.loc[a, "2016 Wealth"] = int(float(x[1:4]))
    
    new3 = new2[new2.find("N"):]
    if "DEBUT" in i:
        new3 = i.replace(" ", "")
        new3 = new3[new3.find("DEBUT")+5:new3.find("AGE:")].replace(",", ", ").replace("FINANCIALSERVICES", "FINANCIAL SERVICES")
    elif "RE-ENTRY" in i:
        new3 = i.replace(" ", "")
        new3 = new3[new3.find("RE-ENTRY")+8:new3.find("AGE:")].replace(",", ", ").replace("FINANCIALSERVICES", "FINANCIAL SERVICES")
    else:
        new3 = new3[1: new3.find("AGE:")]
        if new3.find("AGES:") != -1:
            new3 = new3[: new3.find("AGES:")]
    results.loc[a, "Sector"] = new3
    
    new4 = i
    if "DEBUT" in i:
        new4 = i[i.find("DEBUT")+5:]
    elif "RE-ENTRY" in i:
        new4 = i[i.find("RE-ENTRY")+8:]
    else:
        new4 = new2[new2.find("N"):]
    
    new4 = new4[new4.find("AGE"):]
    new4 = new4[new4.find(":")+2:]
        
    if comma:
        new5 = new4[: new4.find(",", 4, 10)]
    else:
        new5 = new4[: new4.find(",")]
    results.loc[a, "Age"] = new5
    

    if comma:
        new4 = new4[new4.find(",", 4, 10)+2:].strip()
    else:
        new4 = new4[new4.find(",")+2:].strip()
    new = False

    
    for d,e in enumerate(new4):
        if e == ",":
            new = True
        elif new:
            new = False
        elif not (e.isupper()):
            new6 = new4[:d-1]
            new7 = new4[d-1:]
            break
            
            
    results.loc[a,"Location"] = new6
    
    if new7.find("FULL TEXT") != -1:
        new7 = new7[new7.find("FULL TEXT")+9:]
        new7 = new7[new7.find("AGE") +12:]
    if new7.find("PDF GENERATED") != -1:
        new8 = new7[:new7.find("\n\n")]
        new8 = new8[:new8.find("PDF GENERATED")]
        new7 = new7[new7.find("Page"):]
        new7 = new7[new7.find("\n\n")+4:]
        new7 = "".join([new8,new7])
    if new7.find("ABSTRACT (ENGLISH)") != -1:
        new7 = new7[:new7.find("DETAILS")]
        new7 = new7[:new7.find("ProQuest")]
    if new7.find("Publication subject") != -1:
        new7 = new7[:new7.find("Publication subject")]
    if new7.find("PDFGENERATED") != -1:
        new8 = new7[:new7.find("\n\n")]
        new8 = new8[:new8.find("PDFGENERATED")]
        new7 = new7[new7.find("Page"):]
        new7 = new7[new7.find("\n\n")+4:]
    if new7.find("PROQUEST.COM") != -1:
        new8 = new7[:new7.find("\n\n")]
        new8 = new8[:new8.find("PROQUEST.COM")-10]
        new7 = new7[new7.find("Page"):]
        new7 = new7[new7.find("\n\n")+4:]
        
    results.loc[a, "Additional Text"] = new7


    
        
    

results
```
